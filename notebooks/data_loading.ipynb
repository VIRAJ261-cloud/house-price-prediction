# Import Modules

```python
import pandas as pd 
import numpy as np
import os
```

# Define File path

```python
BASE_DIR = os.path.dirname(os.getcwd())
DATA_PATH = os.path.join(BASE_DIR,"data","raw","train.csv")

print(f"Dataset is at:\n{DATA_PATH}")
```

# Load the Dataset

```python
df = pd.read_csv(DATA_PATH)
df = df.iloc[:,1:]

print("Dataset loaded successfully")
print(f"Rows : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")
```

# Columns Names and Data types

```python
print("="*50)
print("Columns Names and Data types")
print("="*50)
print(df.dtypes.to_string())
```

# Sample rows 

```python
print("="*50)
print("RANDOM ROWS OF DATASET")
print("="*50)
df.sample(5)
```

# Statistical Summary 

```python
print("="*50)
print("Statistical Summary of the dataset")
print("="*50)
df.describe()
```

# Check for missing values


```python
missing = df.isnull().sum()
missing_percent = (missing*100)/df.shape[0]
missing_report = pd.DataFrame({
    'Missing Count':missing,
    'Missing %' : missing_percent.round(2)
})
missing_report = missing_report[missing_report['Missing Count']>0]
missing_report = missing_report.sort_values('Missing %',ascending=False)

print('='*50)
print(f"Columns with missing values : {len(missing_report)} out of {df.shape[1]}")
print('='*50)
print(missing_report.to_string())
```

# Check for Duplicate Rows

```python
duplicates = df.duplicated().sum()
print(f"Duplicates rows found: {duplicates}")

if(duplicates==0):
    print("No duplicates , Dataset is clean")
else:
    print(f"Found {duplicates} duplicates , need to remove in preprocessing")
```

# Target Variable Analysis

```python
print("TARGET VARIABLE: SalePrice")
print("="*50)
print(f"Min Price : {df['SalePrice'].min()}")
print(f"Max Price : {df['SalePrice'].max()}")
print(f"Mean Price : {df['SalePrice'].mean():,.0f}")
print(f"Median Price : {df['SalePrice'].median():,.0f}")
print(f"Standard Deviation : {df['SalePrice'].std():,.0f}")
```
